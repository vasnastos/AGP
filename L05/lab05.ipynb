```python
import random as r
from time import time
import termcolor as tm
import numpy.testing as npt
from functools import reduce

r.seed(time()*1000)
e=2.7182

def RandomSummary():
    sumboard=[]
    while sum(sumboard)<1:
        sumboard.append(r.random())
    return len(sumboard)

def checkN(times):
    totalcounter=0
    for x in range(int(times)):
        totalcounter+=RandomSummary()
    totalcounter/=times
    print(abs(totalcounter),abs(e))
    print('Αναμενόμενη Τιμη:'+str(totalcounter),end='-->')
    return abs(totalcounter)-0.05<=abs(e) and abs(totalcounter)+0.05>=abs(e)

array=[10**x for x in range(2,8)]
cols={True:'green',False:'red'}
for x in array:
    out=checkN(x)
    print(tm.colored(str(out),cols[out]))
```

```python
import pickle as pc
import random as ra
import time as t
import math as m

ra.seed(t.time()*1000)

days={1:31,2:29,3:31,4:30,5:31,6:30,7:31,8:31,9:30,10:31,11:30,12:31}

def Paragontical(n):
    if int(n)==1:
        return 0
    sum=1
    for i in range(1,n+1):
        sum*=i
    return sum

def BirthDayParadox(number_of_persons):
    propability=1-(Paragontical(365)/(365**number_of_persons*Paragontical(365-number_of_persons)))
    propability*=100
    return propability

def Paradox_Per_Population():
    population=[5,10,20,23,40,50,60,70,75,100]
    print('Population  \t  Propability')
    for x in population:
        print(f'P:{x}        \t  Prb:{BirthDayParadox(x)}%')


def BirthDay(counter):
    bdays=[]
    for x in range(int(counter)):
        month=ra.randint(1,12)
        bdays.append({ra.randint(1,month):int(month)})
    return bdays

def Paradox(bdays):
  uniquedates={}
  for bday in bdays:
        date=''
        for x in bday:
          date=str(x)+'-'+str(bday[x])
        if date in uniquedates:
            uniquedates[date]+=1
        else:
            uniquedates.update({date:1})
  return uniquedates

def check_Paradox():
    birthdates=BirthDay(100)
    BirthDayParadox(100)
    dates=Paradox(birthdates)
    accept=list(filter(lambda elem:elem[1]>1,dates.items()))
    for x in accept:
        date='0'+str(x[0].split('-')[0]) if int(x[0].split('-')[0])<10 else str(x[0].split('-')[0])
        month='0'+str(x[0].split('-')[1]) if int(x[0].split('-')[1])<10 else str(x[0].split('-')[1])
        print(f'Paradox found-Date:{date}/{month}')
    with open('savefile','wb') as f: 
        pc.dump(dates,f)

check_Paradox()
print('================================================================')
Paradox_Per_Population()
```

```python
import pip._vendor.requests as req
from lxml import html 
import bs4
def Decode(response):
    soup=bs4.BeautifulSoup(response.text,features='lxml')
    table=soup.findAll('table')
    for x in table:
        cells=x.findAll('tr')
        for tr in cells:
            for td in tr.findAll('td'):
               tddec=str(td).replace('<br/>',' ')
               print(html.fromstring(tddec).text_content(),end=' ')
            print(end='\n')

resp=req.get('https://www.dit.uoi.gr/members.php')
Decode(resp)
```
